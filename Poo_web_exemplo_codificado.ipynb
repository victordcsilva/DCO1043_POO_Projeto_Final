## WebCrawler OLX

```python
import requests 
from bs4 import BeautifulSoup

def  OlxSpider(max_pages): #Informação de n paginas
    page = 1
    while page <= max_pages:
        url = 'http://rn.olx.com.br/?o='+str(page)
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text,"lxml") #Todo o codico fonte de uma pagina
        for link in soup.findAll('a', href=True): #todos os links 'a' com tag 'href'
            href=link['href']
            if href[0:40]=='http://rn.olx.com.br/rio-grande-do-norte':
                print("URL encontrado:",href)
                title=link['title']
                print("Titulo do URL:",title)
                Pegar_dados_individualmente(href)
        page = page + 1

def Pegar_dados_individualmente(item_url): #Informação em dada link de uma unica pagina
    source_code = requests.get(item_url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text,"lxml") 
    for item_name in soup.findAll('span',{'class':'OLXad-price'}): #toda informação da classe OLX da tag span
        print("Preço:",item_name.string)
        
        
OlxSpider(1)
```

## WebCrawler previsão de tempo

```python
from bs4 import BeautifulSoup
import requests

estado=input('Estado: ').lower() 
cidade=input('Cidade: ')
cidade=cidade[0].upper()+cidade[1::].lower()

site = 'http://www.tempoagora.com.br/previsao-do-tempo/'+estado+'/'+cidade
print ('>> Site:',site,'\n')
r = requests.get(site)
soup = BeautifulSoup(r.text, "html5lib")

print ('\n======[ Previsao do Tempo: ]======')
for ul in soup.find_all('ul', {'class': 'dsp-table day-info'}):
    for li in ul.find_all('li'):
        print (li.text.replace("\t", " ").replace("\r", "").replace("\n", ""))

print ('\n======[ Proximas horas: ]======')
for ul in soup.find_all('ul', {'class': 'dsp-table next-hours-list'}):
    for li in ul.find_all('li'):
        print (li.text.replace("\t", " ").replace("\r", "").replace("\n", ""))
```
